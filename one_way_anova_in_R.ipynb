# One Way ANOVA

ANOVA is an ANalysis Of VAriance. It is used to determine if two separate samples were derived from the same population. It can also be used to find out if there is a statistical significant difference between two samples or if the difference is caused by random chance. It describes the variance within each group and the variance between groups.
